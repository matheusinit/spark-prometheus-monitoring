```python
from pyspark.sql import SparkSession

spark = SparkSession \
  .builder \
  .appName("Write raw data to parquet") \
  .master("spark://0.0.0.0:7077") \
  .getOrCreate()

df = spark.read.csv('./data/Amazon Sale Report.csv')
```

```python
df.show()
```

```python
from pyspark.sql import Window
from pyspark.sql.functions import rank, desc

window_spec = Window.partitionBy("Status").orderBy(desc("Qty"))
ranked_data = df.withColumn("rank", rank().over(window_spec))
```
